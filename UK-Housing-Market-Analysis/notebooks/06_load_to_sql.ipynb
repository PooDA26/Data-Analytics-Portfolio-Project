```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine(
    "postgresql://postgres:7487@localhost:5432/uk_housing_db"
)
```

```python
dim_date = pd.read_parquet(
    "../data/processed/dim_date.parquet"
)
```

```python
dim_date.columns = dim_date.columns.str.lower()
```

```python
dim_date.to_sql(
    "dim_date",
    engine,
    if_exists="replace",
    index=False
)
```

```python
fact_housing = pd.read_parquet(
    "../data/processed/fact_housing.parquet"
)
```

```python
fact_sample = fact_housing.sample(
    n=500000,
    random_state=42
)
```

```python
fact_sample.columns = fact_sample.columns.str.lower()
```

```python
fact_sample.to_sql(
    "fact_housing",
    engine,
    if_exists="replace",
    index=False,
    chunksize=50000,
    method="multi"
)
```
