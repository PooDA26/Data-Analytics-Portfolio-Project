```python
import pandas as pd
```

```python
df = pd.read_parquet(
    "../data/processed/uk_housing_analysis.parquet"
)
```

```python
dim_date = df[
    [
        "Date_of_Transfer",
        "Year",
        "Month",
        "Quarter",
        "Month_Name"
    ]
].drop_duplicates()
```

```python
dim_date = dim_date.reset_index(
    drop=True
)

dim_date["Date_Key"] = (
    dim_date.index + 1
)
```

```python
dim_date = dim_date[
    [
        "Date_Key",
        "Date_of_Transfer",
        "Year",
        "Month",
        "Quarter",
        "Month_Name"
    ]
]
```

```python
dim_location = df[
    [
        "Postcode",
        "Town_City",
        "District",
        "County"
    ]
].drop_duplicates()
```

```python
dim_location = (
    dim_location
    .reset_index(drop=True)
)

dim_location["Location_Key"] = (
    dim_location.index + 1
)
```

```python
dim_location = dim_location[
    [
        "Location_Key",
        "Postcode",
        "Town_City",
        "District",
        "County"
    ]
]
```

```python
dim_property = df[
    [
        "Property_Type",
        "Property_Type_Name",
        "New_Build"
    ]
].drop_duplicates()
```

```python
dim_property = (
    dim_property
    .reset_index(drop=True)
)

dim_property["Property_Key"] = (
    dim_property.index + 1
)
```

```python
dim_property = dim_property[
    [
        "Property_Key",
        "Property_Type",
        "Property_Type_Name",
        "New_Build"
    ]
]
```

```python
dim_tenure = df[
    ["Tenure"]
].drop_duplicates()
```

```python
dim_tenure = (
    dim_tenure
    .reset_index(drop=True)
)

dim_tenure["Tenure_Key"] = (
    dim_tenure.index + 1
)
```

```python
dim_tenure = dim_tenure[
    [
        "Tenure_Key",
        "Tenure"
    ]
]
```

```python
fact = df.merge(
    dim_date,
    on=[
        "Date_of_Transfer",
        "Year",
        "Month",
        "Quarter",
        "Month_Name"
    ]
)
```

```python
fact = fact.merge(
    dim_location,
    on=[
        "Postcode",
        "Town_City",
        "District",
        "County"
    ]
)
```

```python
fact = fact.merge(
    dim_property,
    on=[
        "Property_Type",
        "Property_Type_Name",
        "New_Build"
    ]
)
```

```python
fact = fact.merge(
    dim_tenure,
    on="Tenure"
)
```

```python
fact_housing = fact[
    [
        "Transaction_ID",
        "Date_Key",
        "Location_Key",
        "Property_Key",
        "Tenure_Key",
        "Price",
        "Source_Year"
    ]
]
```

```python
dim_date.to_parquet(
    "../data/processed/dim_date.parquet",
    index=False
)

dim_location.to_parquet(
    "../data/processed/dim_location.parquet",
    index=False
)

dim_property.to_parquet(
    "../data/processed/dim_property.parquet",
    index=False
)

dim_tenure.to_parquet(
    "../data/processed/dim_tenure.parquet",
    index=False
)

fact_housing.to_parquet(
    "../data/processed/fact_housing.parquet",
    index=False
)
```
