```python
df = df.dropna(
    subset=["Postcode"]
)
```

```python
df.duplicated().sum()
```

```python
df["Date_of_Transfer"] = pd.to_datetime(
    df["Date_of_Transfer"]
)
```

```python
df.info()
```

```python
df["Price"] = pd.to_numeric(
    df["Price"],
    errors="coerce"
)
```

```python
df["Year"] = (
    df["Date_of_Transfer"]
    .dt.year
)
```

```python
df["Month"] = (
    df["Date_of_Transfer"]
    .dt.month
)
```

```python
df["Quarter"] = (
    df["Date_of_Transfer"]
    .dt.quarter
)
```

```python
df["Month_Name"] = (
    df["Date_of_Transfer"]
    .dt.month_name()
)
```

```python
property_map = {
    "D":"Detached",
    "S":"Semi-Detached",
    "T":"Terraced",
    "F":"Flat",
    "O":"Other"
}

df["Property_Type_Name"] = (
    df["Property_Type"]
    .map(property_map)
)

new_build_map = {
    "Y":"Yes",
    "N":"No"
}

df["New_Build"] = (
    df["Old_New"]
    .map(new_build_map)
)

tenure_map = {
    "F":"Freehold",
    "L":"Leasehold"
}

df["Tenure"] = (
    df["Duration"]
    .map(tenure_map)
)
```
